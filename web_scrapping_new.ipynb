```python
def extract_job_urls_from_result(soup): 
    urls = []
    for div in soup.find_all(name = "div", attrs = {"class":"row"}):
        for a in div.find_all(name = "a", attrs = {"data-tn-element":"jobTitle"}):
            this_url = a['href']
            to_go_url = "https://www.indeed.com/viewjob" + this_url[7:]
            urls.append(to_go_url)
    return(urls)
```

```python
whole_data=pd.DataFrame({'Job_Title':[],'Description':[]})
```

```python
for job in job_list:
    Job_Title=[]
    Description=[]
    listing_urls_list = []
    for start in range(0, 620, 10):
        URL="https://www.indeed.co.in/jobs?q="+job.replace(' ', '+')+"&start="+str(start)
        listing_urls_list.append((job, URL))
    jobs_url=[]
    for job,url in listing_urls_list:
        response=requests.get(url)
        page = response.content
        time.sleep(1)
        soup = BeautifulSoup(page,"html.parser", from_encoding="utf-8")
        found_url=extract_job_urls_from_result(soup)
        for u in found_url:
            jobs_url.append((job,u))
    for job,url in jobs_url:
        Job_Title.append(job)
        page = requests.get(url)
        soup = BeautifulSoup(page.text, "html.parser")
        spans = soup.findAll('div', attrs={"class": "jobsearch-jobDescriptionText"})
        for span in spans:
            Description.append(span.text.strip())
    Job_Title=Job_Title[0:600]
    Description=Description[0:600]
    if(len(Description)<600):
        k=600-len(Description)
        Description=Description+Description[0:k]
    scraped_data = pd.DataFrame({'Job_Title':Job_Title, 'Description': Description})
    #scraped_data=scraped_data.assign(job_title=job)
    frames = [whole_data, scraped_data]
    whole_data = pd.concat(frames)
    
```

```python
from nltk import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

import pandas as pd
```

```python
lemmatizer = WordNetLemmatizer()
symbols_to_ignore = ['(', ')', '.', '\\', ':', '%', '’','/','-','&']
to_ignore = stopwords.words('english')
to_ignore.extend(symbols_to_ignore)
to_ignore.extend(extra)
```

```python
extra=['delhi','hyderabad','bangalore','patna','new york','role','responsibilites','manage','management','successful','knowledge','help','skill','professional','salary','disability','join',' ']
```

```python
cleaned_jobs=[]
cleaned_jobs=list(whole_data.Description)
```

```python
for i in range(len(cleaned_jobs)):
    cleaned_jobs[i] = re.sub("[^a-zA-Z]", " ",cleaned_jobs[i])
```

```python
#cleaned_jobs
```

```python
new_cleaned_jobs = []
for job in cleaned_jobs:

	#Convert to lowercase and remove some symbols from the start
    lower_job = job.lower().replace(',', '').replace('_', '').replace(';', '')
    
    #Tokenize into words
    words_1 = word_tokenize(lower_job)

    #Lemmatize the words
    words_2 = [lemmatizer.lemmatize(w) for w in words_1]

    #Skip stopwords and some extra symbols (see line 3)
    words_3 = [w for w in words_2 if not w in to_ignore]

    #Join the tokens into a string and store it
    clean_job = " ".join(words_3)
    new_cleaned_jobs.append(clean_job)
```

```python
cleaned_dataset=pd.DataFrame({'Job_Title':Job_Title,'Description':new_cleaned_jobs})
```
